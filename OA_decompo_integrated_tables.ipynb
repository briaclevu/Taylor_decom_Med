**Load the result of the decomposition**

```python
%%octave
# load the 2D fields for terme of the decomposition
load('/home/biomac1/blevu/SCRATCH/DE9_2000/deconvolution_H_OA_2084_2099.mat')
dirout='/home/biomac1/blevu/SCRATCH/DE9_2000/';
# name for each terms (delta is august - february, D is for the delta calculated, dHd and dOAd are partial for H and OemagaA)
name={'delta','DH','DOA','dHd','dOAd'};
# scale per term for readability of the table
scale=[1 1 1 1 1e3];
mask={'MED' 'BWE' 'BEA' 'ADR' 'AEG'};
```

**Create csv table with terms and part averaged over the MED bassin for all simulations**

```python
%%octave
# create simple .csv per terms
for k=1:length(mask)
for j=1:length(name)
# file and the table title is the same; 1 file per 1 term and 1 area (here is MED)
 filename=[dirout,char(name(j)),'_',char(mask(k)),'.csv'];
 fid = fopen(filename, 'w');
 if j>4
  fprintf(fid,[char(name(j)),'_',char(mask(k)),' x1e3']);
 else
  fprintf(fid,[char(name(j)),'_',char(mask(k)),'']);
 end
# The row are the different simulations: 2b=CTRL; b=CTRL+CO2; 2a=CLIM; a=CLIM+CO2
 for t=1:2
  for d=1:2
   fprintf(fid,[',Simu_',char(CD(d)),char(CT(t))]);
  end
 end
 fprintf(fid,'\n');
# indice of the arameters in field2={'T' 'S' 'PO4' 'Si' 'DIC' 'ALK' 'H' 'OA'};
 if j==1
  inter=[1:2 5:8];
 else
  inter=[1:2 5:6];
 end
# Scan the variable parameter for each term of the decomposition
for i=inter
 fprintf(fid,[char(name(j)),char(field2(i))]); # write the term
 for t=1:2
   for d=1:2
     eval(['fprintf(fid,'',%.2f'',nansum(nansum(',...
          char(name(j)),char(field2(i)),char(CD(d)),char(CT(t)),'.*ds2d.*mask_',char(mask(k)),'))/S_',char(mask(k)),'*scale(j));'])
   end
 end
 fprintf(fid,'\n'); # line jump after written the averaged value of the term over the whole bassin 
end
# Add more terme with DHC and DOAC which are sum of the DIC and ALK terms
if j==2 || j==3
 fprintf(fid,[char(name(j)),'C']);
 for t=1:2
   for d=1:2
     eval(['fprintf(fid,'',%.2f'',nansum(nansum(',...
           char(name(j)),'C',char(CD(d)),char(CT(t)),'.*ds2d.*mask_',char(mask(k)),'))/S_',char(mask(k)),');'])
   end
 end
 fprintf(fid,'\n');
#
 fprintf(fid,char(name(j)));
 for t=1:2
   for d=1:2
     eval(['fprintf(fid,'',%.2f'',nansum(nansum(',...
           char(name(j)),char(CD(d)),char(CT(t)),'.*ds2d.*mask_',char(mask(k)),'))/S_',char(mask(k)),');'])
   end
 end
 fprintf(fid,'\n');
end
fclose(fid);
#
end # loop terms
end # loop area
```

**Sumurazing tables**
--

The change in amplitude of H⁺ in a high CO2 world is almost 100% (see figure ($\Delta H^+ = f (CO_2)$) and is very similar between the CO2 experiment and CO2+clim experiment. WHereas the amplitude of OmegaAr doe not change very much between experiment (+20% due to cimate effect). To explore what are the main drivers of the amplitude in the different simulation and how they change, we calculate the part in the H⁺ and OmegaAr amplitude between february and august ($\Delta H^+$ and $\Delta OA$) for the different parameter in the different simulation.

**$\delta H^+$/ $\delta X$ x $\Delta X$ / $\Delta H$ (%)**

In all simulation the H⁺ amplitude seems mainly driven by the amplitude of temperature ($\Delta T$ 119% to 156% in the table) but $\Delta T$ is partially compensated by by the term due to $\Delta ALK$ and $\Delta DIC$ (ie. $\Delta C$) from 17% to 55% in absolute value of the $\Delta H$. 
legend: Last 16 years of the control experiment (1984-1999) representing the preindustrial conditions; CO2 experiment and CO2 + climatic experiment.
All are average for the whole bassin.
The change changes in these proportions of $\Delta T$ and $\Delta C$ suggest that the CO2 invasion to the ocean has two effect: the most important increases the term of $\Delta T$ through the increase of the partial $\delta H^+$ / $\delta T$ and an other increasing the compensation term of $\Delta C$ through the change in partials of $\delta H^+$ / $\delta DIC$ and $\delta H^+$ / $\delta ALK$. Effect of the climatic change which is an increase of $\delta H^+$ / $\delta T$ is almost fully compensated by the change in $\Delta DIC$ in the Western and Ionian bassiin and by $\Delta ALK$ in the other bassin. 
The term due to $\Delta S$ is never significant but in some area like north Agean Sea with the influence of Dardanelles with strong Alkalinity waters...

**$\delta OA$/ $\delta X$ x $\Delta X$ / $\Delta OA$ (%)**

There is a small change between different simulation in the amplitude of OmegaAr.
In the control and the CO2 experiment, the amplitude is driven 50 percent by $\Delta T$ and 50 percent by $\Delta C$. The decrease of OA amplitude in high CO2 is driven by the decrease of $\Delta T$ due to 
the decrease by 30 % of the partial $\delta H^+$ / $\delta T$. The effect of climate change is the most important with a part due to 15% increase of the partial $\delta H^+$ / $\delta T$ and the main part due to the strong decrease in amplitude of surface DIC in the West and Ioanian bassin and the strong increase in amplitude of surface ALK in the other bassin.
